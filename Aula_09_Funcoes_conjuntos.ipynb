Faça um programa que implemente um funçao que receba como parâmetros um valor inicial (start), um valor final (end) e um número de pontos (n) e devolva uma lista com n pontos contendo valores em Progressão aritmética entre start e end. Chame a função de linspace. Exemplo: linspace(0, 10, 5) deve retornar [0, 2.5, 5.0, 7.5, 10.0]

```python
def linspace(start, end, n):
  if n==1:
    return [start]
  resp = []
  passo = (end-start)/(n-1)
  for i in range(n):
    resp.append(start+passo*i)
  return resp

t = linspace(0,10,5)
print(t)
print(len(t))
```

Faça um função que recebeba uma lista com um conjunto de valores temporais e calcule a função de cinemática: s = so + vo.t + a.t^2 / 2.  

```python
def posicao(s0, v0, a, t):
  s = []
  for x in t:
    s.append(s0+v0*x+a*x**2/2)
  return s

s0 = 2.0
v0 = -20.0
a = 10.0
t = linspace(0,10,100)
s = posicao(s0,v0,a,t)
print(t)
print(s)
```

```python
import matplotlib.pyplot as plt  #carrega a biblioteca gráfica matplotlib
plt.plot(t,s,'red')
plt.title("Posição")
plt.xlabel("tempo (s)")
plt.ylabel("posção (m)")
plt.show()
```

Faça um função que recebeba uma lista com um conjunto de valores temporais e calcule a função de cinemática: v = vo + a.t

```python
def velocidade(v0,a,t):
  v = []
  for x in t:
    v.append(v0+a*x)
  return v

v = velocidade(v0,a,t)
print(v)
```

```python
plt.plot(t,v,'blue')
plt.title("Velocidade")
plt.xlabel("tempo (s)")
plt.ylabel("velocidade (m/s)")
plt.show()
```
